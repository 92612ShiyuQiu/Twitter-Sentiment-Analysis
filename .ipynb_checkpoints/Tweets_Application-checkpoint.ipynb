```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import VotingClassifier
```

```python
X = np.genfromtxt('data/X.csv', delimiter=',')
Y = np.genfromtxt('data/Y.csv', delimiter=',')
tradewar_posts = pd.read_csv('dataset/tradewar.csv')
```

```python
tradewar_posts2 = tradewar_posts.drop(tradewar_posts.columns[0], axis=1)
tradewar_posts_list = tradewar_posts[tradewar_posts2.columns[0]].values.tolist()
tradewar_posts_list
```

```python
trump_posts = pd.read_csv('dataset/trump.csv')
trump_posts2 = trump_posts.drop(trump_posts.columns[0], axis=1)
trump_posts_list = trump_posts[trump_posts2.columns[0]].values.tolist()
trump_posts_list
```

```python
coronavirus_posts = pd.read_csv('dataset/coronavirus.csv')
coronavirus_posts2 = coronavirus_posts.drop(coronavirus_posts.columns[0], axis=1)
coronavirus_posts_list = coronavirus_posts[coronavirus_posts2.columns[0]].values.tolist()
coronavirus_posts_list
```

```python
X = X[1:]
Y = Y[1:]
```

```python
X[:,:-1]
```

```python
X_comp, X_test, Y_comp, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
Xtr, Xva, Ytr, Yva = train_test_split(X_comp, Y_comp, test_size=0.2)
```

```python
lr = LogisticRegression(C=2, penalty='l1')

mlp = MLPClassifier(max_iter=1000, hidden_layer_sizes=(100,), activation='identity', 
                            solver='sgd', alpha=0.01, learning_rate='adaptive')

gb = GradientBoostingClassifier(max_depth=3,loss="deviance",learning_rate=0.1,
                                 n_estimators=1500,min_samples_split=20,min_samples_leaf=9,
                                 max_features="sqrt", subsample=0.5)
```

```python
combined = VotingClassifier(estimators=[('lr', lr), ('nn', mlp), ('gb', gb)],
                         voting='soft', flatten_transform=True) # weights=?

print("training started")
combined.fit(Xtr,Ytr)
print("training finished")
```

```python
tradewar_proba = combined.predict_proba(X_tradewar)
trump_proba = combined.predict_proba(X_trump)
coronavirus_proba = combined.predict_proba(X_coronavirus)
print(tradewar_proba, trump_proba, coronavirus_proba)
```

```python
#tradewar
positive = 0
negative = 0
neutral = 0
positive_posts = []
negative_posts = []
neutrual_posts = []

index = 0
for p in tradewar_proba:
    if p[0]>0.56: 
        positive += 1
        if p[0]>0.7:
            positive_posts.append(tradewar_posts_list[index])
    elif p[0]<0.44:
        negative += 1
        if p[0]<0.3:
            negative_posts.append(tradewar_posts_list[index])
    else:
        neutral += 1
        if p[0]>0.49 and p[0]<0.51:
            neutrual_posts.append(tradewar_posts_list[index])
    index += 1
print((positive/len(tradewar_proba))*100,"% holds a positive attitude towards tradewar")
print((negative/len(tradewar_proba))*100,"% holds a negative attitude towards tradewar")
print((neutral/len(tradewar_proba))*100,"% holds a neutral attitude towards tradewar")
print(positive_posts)
print(negative_posts)
print(neutrual_posts)
```
