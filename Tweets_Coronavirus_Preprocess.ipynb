```python
import re

def remove_numbers(post):
    return re.sub(r'\d+', '', post)
```

```python
X_coronavirus_nonum = [remove_numbers(p) for p in X_coronavirus_lower]
```

```python
import string

def remove_punctuation(post):
    translator = str.maketrans('', '', string.punctuation)
    return post.translate(translator)
```

```python
X_coronavirus_nopunc = [remove_punctuation(p) for p in X_coronavirus_nonum]
```

```python
X_coronavirus_nopunc[:10]
```

```python
def remove_whitespace(post):
    return " ".join(post.split())

X_coronavirus_nospace = [remove_whitespace(p) for p in X_coronavirus_nopunc]
```

```python
X_coronavirus_nospace[:10]
```

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words("english"))

def remove_stopwords(post):
    tokens = word_tokenize(post)
    tokens_nostop = [w for w in tokens if w not in stop_words]
    return tokens_nostop
```

```python
coronavirus_tokens_list = [remove_stopwords(p) for p in X_coronavirus_nospace]
```

```python
coronavirus_tokens_list
```

```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

stemmer = WordNetLemmatizer()
analyzer = CountVectorizer().build_analyzer()

def stemmed_words(post):
    return (stemmer.lemmatize(w, pos='v') for w in analyzer(post) if w not in stop_words)

#vectorizer = CountVectorizer(stop_words='english', min_df=0.005, lowercase=True, decode_error='ignore', analyzer=stemmed_words)
```

```python
all_zero = 0

def get_matrix(features_list, xxx_tokens_list):
    global all_zero
    matrix = []
    for post in xxx_tokens_list:
        l = []
        for f in features_list:
            if f in post:
                l.append(1)
            else:
                l.append(0)
        matrix.append(l)
        if 1 not in l:
            all_zero += 1
    return matrix
```

```python
coronavirus_matrix = get_matrix(features_list, coronavirus_tokens_list)
coronavirus_matrix_df = pd.DataFrame(coronavirus_matrix)
print(all_zero)
```
