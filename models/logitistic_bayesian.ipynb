```python
import numpy as np
from skopt import gp_minimize
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from skopt.space import Real
from skopt.utils import use_named_args
```

```python
X = np.genfromtxt('../data/X.csv', delimiter=',')
Y = np.genfromtxt('../data/Y.csv', delimiter=',')
```

```python
X_comp, X_test, Y_comp, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
Xtr, Xva, Ytr, Yva = train_test_split(X_comp, Y_comp, test_size=0.2)
```

```python
lr = LogisticRegression(penalty='l1')
```

```python
search_space = [Real(0, 10, name="C")]
```

```python
from numpy import mean

@use_named_args(search_space)
def evaluate_model(**params):
    lr.set_params(**params)
    # calculate 5-fold cross validation
    result = cross_val_score(lr, Xtr[:5000], Ytr[:5000], cv=5, verbose=10, n_jobs=-1, scoring='accuracy')
    # calculate the mean of the scores
    estimate = mean(result)
    return 1.0 - estimate
```

```python
result = gp_minimize(evaluate_model, search_space)
print('Best Accuracy: %.3f' % (1.0 - result.fun))
print('Best Parameters: C=%f' % (result.x[0]))
```

```python
from sklearn.metrics import roc_auc_score
classifier = LogisticRegression(C=0.876839, penalty='l1')
classifier.fit(Xtr, Ytr)
```

```python
print(roc_auc_score(Yva, classifier.predict_proba(Xva)[:,1]))
print(classifier.score(Xtr, Ytr))
print(classifier.score(Xva, Yva))
```
